```python
### Split data into train/val/test

import os
import random
import shutil

# Set your data directories
IMAGE_DIR = '/Users/benmartin/Downloads/Jack_data_sep28__2023/obj_train_data/images/images'
ANNOTATION_DIR = '/Users/benmartin/Downloads/Jack_data_sep28__2023/obj_train_data/images/labels'

# Create directories for the splits
os.makedirs('train/images', exist_ok=True)
os.makedirs('train/labels', exist_ok=True)
os.makedirs('val/images', exist_ok=True)
os.makedirs('val/labels', exist_ok=True)
os.makedirs('test/images', exist_ok=True)
os.makedirs('test/labels', exist_ok=True)

# Get all image files
all_images = [f for f in os.listdir(IMAGE_DIR) if f.endswith('.jpg')]

# Shuffle and split
random.shuffle(all_images)
train_split = int(0.8 * len(all_images))
val_split = int(0.9 * len(all_images))

train_images = all_images[:train_split]
val_images = all_images[train_split:val_split]
test_images = all_images[val_split:]

# Helper function to move files
def move_files(file_list, source_img_dir, source_ann_dir, dest_img_dir, dest_ann_dir):
    for filename in file_list:
        shutil.move(os.path.join(source_img_dir, filename), os.path.join(dest_img_dir, filename))
        annotation_file = filename.replace('.jpg', '.txt')
        shutil.move(os.path.join(source_ann_dir, annotation_file), os.path.join(dest_ann_dir, annotation_file))

# Move the files
move_files(train_images, IMAGE_DIR, ANNOTATION_DIR, 'train/images', 'train/labels')
move_files(val_images, IMAGE_DIR, ANNOTATION_DIR, 'val/images', 'val/labels')
move_files(test_images, IMAGE_DIR, ANNOTATION_DIR, 'test/images', 'test/labels')
```
